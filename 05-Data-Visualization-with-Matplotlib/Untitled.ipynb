```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
#functional
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
```

```python
y
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OO
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.2,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Larger plot')
axes2.set_title('smaller plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for curr in axes:
    curr.plot(x,y)
axes[0].set_title('Axis one')
axes[1].set_title('Axis two')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(3,2))
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X squared")
ax.plot(x,x**3,label="X cubed")
ax.set_title('Title')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.legend(loc=10)
```
